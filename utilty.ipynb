```python
import pandas as pd
from tqdm.auto import tqdm

target = ['Dry_Clover_g', 'Dry_Dead_g', 'Dry_Green_g', 'Dry_Total_g', 'GDM_g']
df = pd.read_csv('Data/train.csv')
new_df = pd.DataFrame()
count = 0
for name in tqdm(df['image_path'].unique().tolist()):
    data = df[df['image_path'] == name].reset_index(drop=True)
    new_df.loc[count, 'state'] = data.loc[0, 'state']
    new_df.loc[count, 'species'] = data.loc[0, 'species']
    new_df.loc[count, 'Sampling_Date'] = data.loc[0, 'Sampling_Date']
    new_df.loc[count, 'image_path'] = '../Data/' + name
    for t in target:
        new_df.loc[count, t] = data.loc[data['target_name']==t, 'target'].values[0]
    count+=1


from sklearn.model_selection import KFold, StratifiedKFold, StratifiedGroupKFold
kf= KFold(n_splits=5, random_state=123, shuffle=True)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
sgkf = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=42)
for i, (train_index, test_index) in enumerate(kf.split(new_df['image_path'], new_df['image_path'])):
    for indx in test_index:
        new_df.loc[indx, 'fold']= i

new_df.to_csv('Data/train_sgkf.csv', index=False)
new_df
```
